```python
import pandas as pd
import glob
import os
from google.colab import files

# Upload the file manually
uploaded = files.upload()

# List the uploaded files
import pandas as pd
df = pd.read_csv("distinct_descriptions.csv")

# Preprocess  text data
# Remove missing or empty descriptions
df = df.dropna(subset=['Job Description'])
df = df[df['Job Description'].str.strip() != ""]
```

```python
from bertopic import BERTopic
from umap import UMAP

#Initialize and Fit BERTopic Model
# Extract descriptions for modeling
descriptions = df['Job Description'].tolist()

# Initialize BERTopic
umap_model = UMAP(n_neighbors=15, n_components=5, min_dist=0.0, metric='cosine')
topic_model = BERTopic(language="english", umap_model=umap_model)

# Fit the model on the descriptions
topics, probs = topic_model.fit_transform(descriptions)

# Add the generated topics to the DataFrame
df['Topic'] = topics
```

```python
# Function to split rows where '\n' occurs
def split_rows_on_token(df, column_name, token="\n"):
    # Split rows and expand into multiple rows
    df_split = df[column_name].str.split(token, expand=True).stack().reset_index(level=1, drop=True)
    # Return a new DataFrame with updated rows
    return df.drop(columns=[column_name]).join(df_split.rename(column_name))

# Apply the function
docs = split_rows_on_token(df, column_name='Job Description')

docs_descr = docs['Job Description'].tolist()
```

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

sentence_model = SentenceTransformer("all-MiniLM-L6-v2")  # Example of a Sentence-BERT model

# Encode documents with Sentence-BERT
embeddings = sentence_model.encode(docs_descr, show_progress_bar=True)

# Pass Sentence-BERT embeddings into BERTopic
topic_model = BERTopic()
topics, probs = topic_model.fit_transform(docs_descr, embeddings)
```
